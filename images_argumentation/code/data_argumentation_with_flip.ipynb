```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
path = '../images/dry/'
path_01 = '../images/dry_argumentation/'

image_list = os.listdir(path)
image_list_01 = []

for i in range(len(image_list)):
    sindex = image_list[i].find(".")
    image_list_01.append(image_list[i][:sindex])

for i in range(len(image_list)):
    img = cv2.imread(path+image_list[i])
#     img = image_list[i]
#     img_test = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    img1 = cv2.flip(img, 1)
    img2 = cv2.flip(img, 0)
    img3 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    img4 = cv2.rotate(img3, cv2.ROTATE_90_CLOCKWISE)
    img5 = cv2.rotate(img4, cv2.ROTATE_90_CLOCKWISE)
    img6 = cv2.flip(img3, 1)
    img7 = cv2.flip(img5, 1)
    
    image_list_test = [img1,img2,img3,img4,img5,img6,img7]
    
    for j in range(len(image_list_test)):
        cv2.imwrite(os.path.join(path_01 ,f'{image_list_01[i]}_{j}.jpg'), image_list_test[j])
```
