```python
for c in range(1,100):
    for a in range(1,c):
        b = (c**2 - a**2)**0.5
        if b == int(b):
            print(c,a,b,a+b+c)
```

```python
for c in range(1,100):
    for a in range(1,int(c/2**0.5)+1):
        b=(c**2 - a**2)**0.5
        if b == int(b):
            print(c,a,b,a+b+c)
```

```python
result=[0]*300
for c in range(1,100):
    for a in range(1, int(c/2**0.5)):
        b=(c**2 - a**2)**0.5
        if b == int(b) and a+b+c <= 1000:
            result[int(a+b+c)] += 1
print(result)
```
